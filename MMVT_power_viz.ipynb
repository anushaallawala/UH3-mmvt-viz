```python
import os
import os.path as op
import numpy as np

MMVT_ROOT = '/Users/anushaallawala/mmvt_root'#change this to your local MMVT directory 
MMVT_CODE = op.join(MMVT_ROOT, 'mmvt_code') #don't change this 
MMVT_DIR = op.join(MMVT_ROOT, 'mmvt_blend') #don't change this 
SUBJ = 'DBSTRD001' # subject name 

os.chdir(MMVT_CODE)
from src.utils import utils #import of utilities should only be after changing directory for mmvt code 
# src is the source folder in the mmvt local folder 
```

```python
testcsv = '/Users/anushaallawala/Downloads/powervals_test.csv' #your file containing power values 
#NOTE about this test file - as a sanity check, 
# the values for the first two conditions (prestim and poststim) are the same, while VCVS and SCC are different)

#initialize. 
electrode_names = [] 
values = [] 

#append values for each condition in a for loop 
for line in utils.csv_file_reader(testcsv, ',', 1): 
    print(line)
    electrode_name = line[0]
    prestim = float(line[1])
    stim = float(line[2])
    VCVS = float(line[3])
    SCC = float(line[4])
    electrode_names.append(electrode_name)
    values.append([prestim, stim, VCVS, SCC])
```

```python

#define condition name -  this name will show up as an option in mmvt under the Electrodes panel 
conditions_names = ['prestim','stim','VCVS','SCC']
print(conditions_names)
```

```python
#set variables for output file name 
electrodes_folder = utils.make_dir(op.join(MMVT_DIR, SUBJ, 'electrodes')) 
outputmmvtfilename = op.join(electrodes_folder, 'electrodes_data_test.npz') # need it to start with electrodes_data
print(outputmmvtfilename)
```

```python
#save file to the mmvt directory defined earlier. 
np.savez(outputmmvtfilename, data=values, names=electrode_names, conditions=conditions_names) #z means we are going to save dictionary of variables (like storing many vars in mat file)
```

```python
#Load the .npz file we just made and check that it looks correct. 
data=np.load(outputmmvtfilename)
print(data['names'], data['conditions'], data['data']) #this should print the electrode names, experiment conditions and data values. 
```
